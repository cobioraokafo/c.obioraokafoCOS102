```python
import tkinter as tk
from tkinter import messagebox

# Global variable to store the total price
total_price = 0

# Define food category and their items
categories = [
    FoodCategory("RICE/PASTA", {"Jollof Rice": 350, "Coconut Fried Rice": 350, "Jollof Spaghetti": 350}),
    FoodCategory("PROTEINS", {"Sweet Chili Chicken": 1100, "Grilled Chicken Wings": 400, "Fried Beef": 400, "Fried Fish": 500, "Boiled Egg": 200, "Sauteed Sausages": 200}),
    FoodCategory("SIDE DISHES", {"Savoury Beans": 350, "Rosted Sweet Potatoes": 300, "Fried Plantains": 150, "Mixed Vegetable Salad": 150, "Boiled Yam": 150}),
    FoodCategory("SOUPS AND SWALLOWS", {"Eba": 100, "Poundo Yam": 100, "Semo": 100, "Atama Soup": 450, "Egusi Soup": 480}),
    FoodCategory("BEVERAGES", {"Water": 200, "Glass Drink(35cl)": 150, "PET Drink(35cl)": 300, "PET Drink(50cl)": 350, "Glass/Canned Malt": 500, "Fresh Yo": 600, "Pineapple Juice": 350, "Mango Juice": 350, "Zobo Drink": 350})
]


class FoodCategory:
    def __init__(self, name, items):
        self.name = name
        self.items = items

    def select_item(self, item_price):
        quantity_window = tk.Toplevel()
        quantity_window.geometry("600x200")
        quantity_label = tk.Label(quantity_window, text="Please enter the amount of portions:")
        quantity_label.pack()
        quantity_entry = tk.Entry(quantity_window)
        quantity_entry.pack()
        next_button = tk.Button(quantity_window, text="Next", command=lambda: self.cashout(quantity_entry.get(), item_price))
        next_button.pack()

    def cashout(self, quantity, item_price):
        global total_price
        total_price += self.calculate_total_price(int(quantity), item_price)
        messagebox.showinfo("Item added to cart", "Close this window to continue")
        cashbox = tk.Toplevel()
        cashbox.geometry("300x100")
        message_label = tk.Label(cashbox, text="Do you wish to checkout")
        message_label.pack()
        yes_button = tk.Button(cashbox, text="Yes", command=lambda: self.yes_option(cashbox))
        yes_button.pack()
        no_button = tk.Button(cashbox, text="No", command=lambda: self.no_option(cashbox))
        no_button.pack()

    def no_option(self, box):
        box.destroy()
        PAUcaf.deiconify()

    def yes_option(self, box):
        global total_price
        box.destroy()
        messagebox.showinfo("", f"{name_entry.get()}, the total cost of your food is â‚¦{total_price}")
        # Reset total price for the next transaction
        total_price = 0

    def calculate_total_price(self, quantity, price):
        total = quantity * price
        if total < 2500:
            total_price = total * (1 - 0.1)
        elif 2500 <= total < 5000:
            total_price = total * (1 - 0.15)
        else:
            total_price = total * (1 - 0.25)
        return total_price

#Main Window    
PAUcaf = tk.Tk()
PAUcaf.title("Welcome to PAU Cafeteria")
PAUcaf.geometry("600x200")

#create menu options
name_label = tk.Label(PAUcaf, text="Please enter your name")
name_label.pack()
name_entry = tk.Entry(PAUcaf)
name_entry.pack()

menu_label = tk.Label(PAUcaf, text="Please select your desired category")
menu_label.pack()


# Create buttons for each category
for category in categories:
    button = tk.Button(PAUcaf, text=category.name, command=category.select_item)
    button.pack()

PAUcaf.mainloop()
```
